```python
def Function(a, b, c):
    if c == "+":
        print(a + b)
    elif c == "-":
        print(a - b)
    elif c == "*":
        print(a * b)
    elif c == "/":
        print(a // b, a % b)
Function(2, 3, "/")
```

```python
import numpy
import pandas
import matplotlib.pyplot
array = numpy.random.randint(0, 10, (10, 10))
number = []
print(array)
for i in range(10):
    count = numpy.count_nonzero(array == i)
    number.append(count)
for i in range(len(number)):
    number_data = pandas.DataFrame({"개수": number})
print(number_data)
for i in range(len(number)):
    matplotlib.pyplot.bar(i, number[i])
matplotlib.pyplot.show()
```

```python
import random
import sys
random_num = []
order_num = []
for x in range(1000):
    random_num.append(random.randint(0, sys.maxsize))
while True:
    random_minimum = random_num[0]
    for x in range(1, len(random_num)):
        if random_minimum > random_num[x]:
            random_minimum = random_num[x]
    order_num.append(random_minimum)
    random_num.remove(random_minimum)
    if random_num == []:
        break  
print(order_num)
```
